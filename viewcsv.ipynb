```python
# Setup
import pandas as pd
from pathlib import Path
```

```python
# Create a list to contain the three CSV files
files = ["Resources/JC-202002-citibike-tripdata.csv/JC-202002-citibike-tripdata.csv",
         "Resources/JC-202003-citibike-tripdata.csv/JC-202003-citibike-tripdata.csv",
         "Resources/JC-202004-citibike-tripdata.csv/JC-202004-citibike-tripdata.csv"]
```

```python
# Create an empty dataframe to store all data
citibike_df = pd.DataFrame()
```

```python
# Concatonate all files into one dataframe
for file in files:
    data = pd.read_csv(file)
    citibike_df = pd.concat([citibike_df, data], axis = 0)
```

```python
# Display dataframe
citibike_df
```

```python
#Store the new dataframe as a CSV file
citibike_df.to_csv('Resources/2022-citibike.csv', index = False)
```
